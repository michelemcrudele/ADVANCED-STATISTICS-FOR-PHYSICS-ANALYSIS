# Exercise 1

```r
compute_probabilities <- function(color, prior, balls, boxes) {
    
    i <- 0:balls
    p_white <- i/balls
    
    if (color=="w") {
        
        N <- sum(prior*p_white)
        prior <- prior*p_white/N
        
    }
    
    if (color=="b") {
        
        N <- sum(prior*(1-p_white))
        prior <- prior*(1-p_white)/N
        
    }
    step <- list(prior, N)
    return (step)
    
}
```

```r
plots <- function(df) {
    
    par(mfrow=c(2,3))
    options(repr.plot.width = 20, repr.plot.height = 10)
    
    for (i in 1:6) {
        
        plot(df[[i]], ylim=c(0,1), type = "p", col="darkblue", pch = 16, cex = 2, ylab = "Probability", 
             cex.lab=1.8, cex.axis = 1.6,cex.main = 2, xlab = "Extraction", main = paste0("BOX:H", i-1))
        
    }
    
}
```

```r
color = ""

balls <- 5
boxes <- 6

i <- 0:balls
p_white <- i/balls

prior <- rep(1/boxes, boxes)

df <- data.frame(matrix(ncol=7, nrow=0))

N <-sum(prior)

df <- rbind(df, matrix(c(prior, N), nrow=1))

while (color != "s") {
    
# Un-comment the following line in order to make the plots appear at every step of the loop. 
# It works only in R terminal!
    
    # plots(df) 
    
    color = readline(prompt = "b for black, w for white; s to stop: ")
    
    if (color=="s") {
        Sys.sleep(0.01)
        print("Simulation Stopped")
        flush.console()
        break
    }
    
    if (color!="b" & color!="w") {
        Sys.sleep(0.01)
        print("Argomento non valido, riprova")
        flush.console()
        next
    }
    
    step <- compute_probabilities(color, prior, balls, boxes)
    prior <- step[[1]]
    N <- step[[2]]
    
    df <- rbind(df, c(prior, N))
    
    Sys.sleep(0.1)    
    cat("P(H0) =", prior[1], "; P(H1) =", prior[2], "; P(H2) =", prior[3],
        "P(H3) =", prior[4], "; P(H4) =", prior[5], "; P(H5) =", prior[6] , "\n")
    flush.console()
    
}

colnames(df) <- c("P(H0)", "P(H1)", "P(H2)", "P(H3)", "P(H4)", "P(H5)", "1/Evidence")
plots(df)
```

```r
box <- sample(0:5, 1)
h <- paste0("H", box)
cat("\nSelected Box =", h, "\n\n")

color = ""

balls <- 5
boxes <- 6

i <- 0:balls
p_white <- i/balls
prior <- rep(1/boxes, boxes)
df <- data.frame(matrix(ncol=7, nrow=0))
N <- sum(prior)
df <- rbind(df, matrix(c(prior, N), nrow=1))


for (i in 0:60) {
    
    color="b"
    
    if (runif(1)<=box/balls) {
        
        color="w"
        
    }
    
    step <- compute_probabilities(color, prior, balls, boxes)
    prior <- step[[1]]
    N <- step[[2]]
    
    df <- rbind(df, c(prior, N))
    
    Sys.sleep(0.01)
    cat(color, '\n')
    cat("P(H0) =", prior[1], "; P(H1) =", prior[2], "; P(H2) =", prior[3],
        "P(H3) =", prior[4], "; P(H4) =", prior[5], "; P(H5) =", prior[6] , "\n")
    flush.console()
    
}

colnames(df) <- c("P(H0)", "P(H1)", "P(H2)", "P(H3)", "P(H4)", "P(H5)", "1/Evidence")
plots(df)
```

```r
gamma <- function(m, beta, n) {
    x = c()
    for (i in 1:n) {
        u <- runif(m)
        x <- c(x, sum(-1/beta*log(u)))
    }
return(x)}
```

```r
g <- gamma(5, 6, 10000)
h <- hist(g, prob=TRUE, col = "darkblue", density=50, xlab="X",
          main="Sampled Erlang distribution points and comparison with theoretical curve")
x <- seq(0, 3, 0.01)
lines(x, dgamma(x, 5, 6), lwd=5, col='red')
```
